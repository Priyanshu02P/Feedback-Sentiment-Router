```python
import re

def is_gibberish(word):
    word = word.lower()

    if len(word) < 3:
        return False

    # Long consonant sequences
    if re.search(r"[bcdfghjklmnpqrstvwxyz]{5,}", word):
        return True

    # Too few vowels
    vowels = sum(c in "aeiou" for c in word)
    if vowels / len(word) < 0.2:
        return True

    return False
```

```python
from itertools import groupby

def remove_consecutive_duplicates(lines, keep=1):
    for _, group in groupby(lines):
        group = list(group)
        group = [g for g in group if not is_gibberish(g)]
        yield from group[:keep]
```

```python
print("".join(remove_consecutive_duplicates(text.splitlines(True))).strip())
```

```python
from presidio_analyzer import AnalyzerEngine
from presidio_anonymizer import AnonymizerEngine

analyzer = AnalyzerEngine()
anonymizer = AnonymizerEngine()

text = """
John Smith lives in New York.
Email: john@gmail.com
Phone: 9876543210
"""

results = analyzer.analyze(
    text=text,
    language="en"
)

result = anonymizer.anonymize(
    text=text,
    analyzer_results=results
)

print(result.text)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def remove_entities(text):
    doc = nlp(text)

    new_text = text
    for ent in reversed(doc.ents):
        if ent.label_ in ["PERSON", "ORG", "GPE", "LOC"]:
            new_text = (
                new_text[:ent.start_char]
                + f"[{ent.label_}]"
                + new_text[ent.end_char:]
            )

    return new_text


text = "John Smith works at Google in California."
print(remove_entities(text))
```

```python
import re

PATTERNS = {
    "EMAIL": r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b",
    "PHONE": r"\b(?:\+91[- ]?)?[6-9]\d{9}\b",
    "IP": r"\b(?:\d{1,3}\.){3}\d{1,3}\b",
    "URL": r"https?://\S+|www\.\S+",
    "AADHAAR": r"\b\d{4}\s?\d{4}\s?\d{4}\b",
    "PAN": r"\b[A-Z]{5}[0-9]{4}[A-Z]\b",
    "CREDIT_CARD": r"\b(?:\d[ -]*?){13,16}\b",
}

def remove_pii(text):
    for tag, pattern in PATTERNS.items():
        text = re.sub(pattern, f"[{tag}]", text)
    return text


sample = """
My email is john.doe@gmail.com.
Phone: +91 9876543210
PAN: ABCDE1234F
Aadhaar: 1234 5678 9012
Website: https://example.com
IP: 192.168.1.1
"""

print(remove_pii(sample))
```
